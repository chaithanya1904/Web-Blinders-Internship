#### Adding New Column

```python
df['Bonus'] = df['Salary'] * 0.1
```

```python
df
```

#### Drop column

```python
df.drop('Bonus',axis=1,inplace=True)
```

```python
df
```

#### rename column names

```python
df.rename(columns={'Experience_Years':'Experience'})
```

```python
df
```

#### Grouping Data

```python
#find avg salary fo each dept
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department').agg(avg_salary=('Salary','mean'))
```

```python
#find avg salary fo each dept and age
df.groupby(['Department','Age']).agg(avg_salary=("Salary",'mean'))
```

```python
#find average experience of employees in each dept
df.groupby('Department').agg(avg_experience=("Experience_Years","mean"))
```

#### sorting data

```python
df_sorted = df.sort_values(by='Salary',ascending=False,inplace=False)
```

```python
df_sorted
```

```python
df_sorted.sort_index(axis=0,ascending=True)
```

#### add bonus of 5% to all employees

```python
df["Salary"] = df["Salary"].apply(lambda x : x+x*0.05)
```

#### Extract the data to csv file

```python
df.sort_values(by="Salary",ascending=False).to_csv('employees_final.csv',index=False)
```
